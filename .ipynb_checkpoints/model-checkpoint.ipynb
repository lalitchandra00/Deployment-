```python
df = df[[col for col in df.columns if col != "Rent"] + ["Rent"]]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
from sklearn.preprocessing import LabelEncoder

area_encoder = LabelEncoder()
city_encoder = LabelEncoder()
furnishing_encoder = LabelEncoder()

df['Area Type'] = area_encoder.fit_transform(df['Area Type'])
df['City'] = city_encoder.fit_transform(df['City'])
df['Furnishing Status'] = furnishing_encoder.fit_transform(df['Furnishing Status'])
```

```python
df.head(2)
```

```python
X = df.iloc[:,0:6]
Y = df["Rent"]
```

```python
from sklearn.decomposition import PCA    
pca = PCA(n_components=1) 
X_reduced = pca.fit_transform(X)
X_1d = X_reduced.ravel()
X_1d = X_reduced.flatten()
X_1d = X_reduced[:, 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 20, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model= RandomForestRegressor(n_estimators=100, max_depth=13)
model.fit(X_train,Y_train)
```

```python
print(model.score(X_train,Y_train)*100)
print(model.score(X_test,Y_test)*100)
```

```python
import pickle
pickle_out = open("model.pkl", "wb")
pickle.dump(model, pickle_out)
pickle_out.close()
```

```python
from sklearn.preprocessing import LabelEncoder

area_encoder = LabelEncoder()
city_encoder = LabelEncoder()
furnishing_encoder = LabelEncoder()

df['Area Type'] = area_encoder.fit_transform(df['Area'])
df['City'] = city_encoder.fit_transform(df['City'])
df['Furnishing'] = furnishing_encoder.fit_transform(df['Furnishing'])


encoders = {
    "Area": area_encoder,
    "City": city_encoder,
    "Furnishing": furnishing_encoder
}

with open("encoders.pkl", "ab") as f:
    pickle.dump(encoders, f)
```
